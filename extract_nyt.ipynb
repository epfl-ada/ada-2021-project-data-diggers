```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import figure
import matplotlib.dates as mdates
import bz2
import json
import os
```

```python
# Extract all quotes from New York Times only
def get_NYT (PATH_IN, PATH_OUT):
  with bz2.open(PATH_IN, 'rb') as file_in:
    with bz2.open(PATH_OUT, 'w') as file_out:
        for instance in file_in:
          instance = json.loads(instance) # loading a sample
          if any(".nytimes.com" in url for url in instance['urls']):
            file_out.write((json.dumps(instance)+'\n').encode('utf-8'))
```

```python
# This has been done separately for each year
PATH_IN = '/content/drive/MyDrive/ADA/Quotebank/quotes-2018.json.bz2'
PATH_OUT = '/content/drive/MyDrive/ADA/NYT/quotes-2018-nyt.json'

get_NYT(PATH_IN, PATH_OUT)
```

```python
DIR_NYT = '/content/drive/MyDrive/ADA/NYT/'
files = os.listdir(DIR_NYT)
df = pd.DataFrame()

# Concatenate the quotes of all years into a single df
for f in files:
  PATH_FILE = DIR_NYT + f
  with bz2.open(PATH_FILE,'rb') as file:
    data = list(map(json.loads, file.readlines()))
    next_df = pd.DataFrame(data)
  df = pd.concat([df, next_df], ignore_index=True)
```

```python
NYT_QUOTES = '/content/drive/MyDrive/ADA/NYT/quotes_nyt.json'
```

```python
# Save all quotes from NYT in a single json file
df.to_json(NYT_QUOTES, orient="index")
```
