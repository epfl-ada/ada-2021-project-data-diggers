##2. Preprocessing of speakers

```python
df_speakers.head()
```

```python
df_speakers['date_of_birth'] = df_speakers['date_of_birth'].apply(lambda date: int(date[0][1:5]) if date is not None and len(date) ==1 else None)
```

```python
df_speakers['date_of_birth'] = df_speakers['date_of_birth'].apply(lambda date: date[0] if date is not None and len(date) ==1 else None)
```

```python
df_speakers[df_speakers['date_of_birth'] == '+2021-00-00T00:00:00Z']
```

```python
df_speakers['nb_aliases'] = df_speakers['aliases'].apply(lambda alias: len(alias) if alias is not None else 0)
```
